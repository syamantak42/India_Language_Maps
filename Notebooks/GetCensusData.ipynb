```python
def fix_dialects(df):    
    df.assign(new_col = "LangGroup")
    df.assign(new_col ="LangGroupCode") 
    df.loc[(df["LangCode"].values)%1000 !=0, "LangGroupCode"] = (df["LangCode"]/1000).astype(int)*1000
    df.loc[np.isnan(df["LangGroupCode"]), "LangGroup"] = df["LangName"]
    df.loc[np.isnan(df["LangGroupCode"]), "LangGroupCode"] = df["LangCode"]
    df = df.sort_values(by = [ "Tehsil", "District", "State", "LangCode"])
    df["LangGroup"] = df["LangGroup"].fillna(method='ffill')
    df = df[['State', 'District', 'Tehsil', 'AreaName', 'LangGroup', 'LangGroupCode', 'LangName','LangCode', 
       'Total', 'Rural', 'Urban']]
    df = df[(df["LangCode"].values)%1000 !=0]
    return df
```

```python
newdf11 = fix_dialects(df11)
newdf01 = fix_dialects(df01)
```

#### Columns have spaces and strings, and duplicates....

```python
for df in [newdf01, newdf11]:
    df['LangGroup'] = df['LangGroup'].str.replace('\d+', '', regex=True).str.replace(' ', '')   
    df['AreaName'] = df['AreaName'].str.replace('\d+', '', regex=True).str.replace("State -", "")\
    .str.replace("District -", "").str.replace(".", "").str.strip()
    
    df.loc[df["LangName"].astype(str).str.contains("Others"), "LangName"] = df["LangGroup"].astype(str)+"_Other"
    df['LangName'] = df['LangName'].str.replace('\d+', '', regex=True).str.replace(' ', '')  
    display(df.head())
```

```python
newdf11= newdf11.drop_duplicates(subset = ["State", "AreaName", "LangName"])
newdf01= newdf01.drop_duplicates(subset = ["State", "AreaName", "LangName"])
#print(newdf11.shape, newdf01.shape, d2.shape)
```

#### Make sure that languages and state names match

```python
for col in ["LangGroup", "LangName"]:
    for d1 in [newdf01, newdf11]:
        for d2 in [newdf01, newdf11]:
            Ldiff = list(set(d1[col].unique().tolist()) - set(d2[col].unique().tolist()))
            Ldiff.sort()
            print( "\n", col, " : ",  Ldiff )
```

```python
for df in [newdf01, newdf11]:
    for Lcol in ["LangGroup", "LangName"]:
        df[Lcol] = df[Lcol].str.replace('\d+', '', regex=True)
        df[Lcol] = df[Lcol].str.replace(' / ', '/', regex=True).str.replace(' /', '/', regex=True).str.replace('/ ', '/', regex=True).str.strip()        
        df[Lcol] = df[Lcol].replace(r'\s+', '_', regex=True)
        df[Lcol] = df[Lcol].str.replace('Oriya', 'Odia')        
    
        df[Lcol] = df[Lcol].str.replace('Bundeli/Bundel_Khandi',  'Bundeli/Bundelkhandi')
        df[Lcol] = df[Lcol].str.replace('Hindi_Multani', 'Multani') 
        
for df in [newdf01]:
    for Lcol in ["LangName"]:     
        df[Lcol] = df[Lcol].str.replace('Chambeali', 'Chambeali/Chamrali')
        df[Lcol] = df[Lcol].str.replace('Bodo', 'Bodo/Boro')
        df[Lcol] = df[Lcol].str.replace( 'Harauti', 'Hara/Harauti')
        df[Lcol] = df[Lcol].str.replace( 'Kok_Barak', 'Kokbarak')
        df[Lcol] = df[Lcol].str.replace('Gangte_Other', 'Gangte')
        df[Lcol] = df[Lcol].str.replace(  'Ho_Other', 'Ho')
        df[Lcol] = df[Lcol].str.replace( 'Karbi/Mikir_Other', 'Karbi/Mikir')
        df[Lcol] = df[Lcol].str.replace('Khond/Kondh_Other', 'Khond/Kondh')
        df[Lcol] = df[Lcol].str.replace('Ladakhi_Other', 'Ladakhi')
        df[Lcol] = df[Lcol].str.replace(  'Monpa_Other', 'Monpa')
        df[Lcol] = df[Lcol].str.replace( 'Miri/Mishing_Other', 'Miri/Mishing')
        
        df[Lcol] = df[Lcol].str.replace('Gojri','Gojri/Gujjari/Gujar')
        df[Lcol] = df[Lcol].str.replace('Proja', 'Proja_(Ori)')
        #df[Lcol] = df[Lcol].str.replace('Bodo_Other', 'Bodo/Boro_Other')
        
        df.loc[df[Lcol] =='Nissi', Lcol ] = 'Nissi/Dafla'        
        df.loc[(df[Lcol] == "Maria")| (df[Lcol] == "Muria"), Lcol]='Maria/Muria'  
        df.loc[df[Lcol] =='Karbi', Lcol ] = 'Karbi/Mikir' 
        
for df in [newdf11]:
    for Lcol in ["LangName"]:        
        df[Lcol] = df[Lcol].str.replace('Afghani/Kabuli/Pashto_Other', 'Afghani/Kabuli/Pashto')
        df[Lcol] = df[Lcol].str.replace('Bodo_Other', 'Bodo/Boro_Other')
        df[Lcol] = df[Lcol].str.replace( 'Jaunpuri/Jaunsari', 'Jaunsari')
        df[Lcol] = df[Lcol].str.replace( 'Lakher_Other', 'Lakher')
        
        
#        df[Lcol] = df[Lcol].str.replace('Proja', 'Proja_(Ori)')#
```
